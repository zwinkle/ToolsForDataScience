```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['MEDV'] = california.target
```

```python
# Boxplot for Median Value (MEDV)
plt.figure(figsize=(8, 5))
sns.boxplot(y=df['MEDV'])
plt.title('Boxplot for Median Value of Owner-Occupied Homes')
plt.ylabel('MEDV')
plt.show()
```

```python
# Barplot for Average Occupancy (AveOccup)
plt.figure(figsize=(6, 4))
sns.countplot(x=pd.cut(df['AveOccup'], bins=[0, 2, 4, 6, 10, 50], labels=['0-2', '2-4', '4-6', '6-10', '10+']))
plt.title('Barplot for Average Occupancy (AveOccup)')
plt.xlabel('AveOccup (Average Occupancy)')
plt.ylabel('Number of Districts')
plt.show()
```

```python
# Boxplot for MEDV vs HouseAge
plt.figure(figsize=(8, 5))
sns.boxplot(x=pd.qcut(df['HouseAge'], q=3, labels=['Young', 'Medium', 'Old']), y=df['MEDV'])
plt.title('Boxplot for MEDV vs HouseAge')
plt.xlabel('House Age Category')
plt.ylabel('MEDV')
plt.show()
```

```python
# Scatter Plot for Latitude vs Longitude
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Longitude'], y=df['Latitude'], hue=df['MEDV'], palette='viridis')
plt.title('Scatter Plot Latitude vs Longitude (Color by MEDV)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Histogram for Median Income
plt.figure(figsize=(8, 5))
sns.histplot(df['MedInc'], bins=20, kde=True, color='blue')
plt.title('Histogram for Median Income')
plt.xlabel('MedInc (Median Income)')
plt.ylabel('Frequency')
plt.show()
```
